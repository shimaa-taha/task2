```python
from pyopenms import *
import pandas as pd
from urllib.request import urlretrieve
url = "https://raw.githubusercontent.com/OpenMS/pyopenms-docs/master/src/data/"
urlretrieve(url+'BSA1.mzML', 'BSA1.mzML')
exp = MSExperiment()
MzMLFile().load('BSA1.mzML', exp)

retention_times = []
mz = []
Intensity = []
count=0
for spec in exp:
    if(count==500):
        break
    if(spec.getMSLevel() == 1):
        MZ,Int=spec.get_peaks()
        mz.extend(MZ)
        Intensity.extend(Int)
        rt=[spec.getRT()] * len(MZ)
        retention_times.extend(rt)
    count=count+1  
```

```python
data = {"MZ": mz,  "Intensity": Intensity, "RT": retention_times}
df = pd.DataFrame(data, columns = ["MZ", "Intensity","RT"])
print(df) 
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.bar(retention_times,Intensity,zs=mz,zdir='x',color='green')
```

```python
rtSpecificMz=[]
initSpecificMz=[]

for i in range(len(mz)):
        if(int(mz[i]) ==500):
            initSpecificMz.append(Intensity[i])
            rtSpecificMz.append(retention_times[i])
```

```python
import matplotlib.pyplot as plt
plt.plot(rtSpecificMz,initSpecificMz,color='green')
plt.xlabel("Retenton time")
plt.ylabel("Intensity")
plt.title("For specific MZ")
plt.show()
```
